### 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot

```python
sns.barplot(x=average_price['fuelType'],y=average_price['price'],hue=average_price['gearbox'])
plt.title('Average price of vehicle by fuel type and gearbox type without andere fuel Type')
plt.xticks(rotation=45)
plt.show()
```

```python
average_price_1 = average_prices[average_prices['fuelType'] == 'andere']
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x=average_price_1['fuelType'],y=average_price_1['price'],hue=average_price_1['gearbox'])
plt.title('Average price of vehicle by fuel type and gearbox type for andere fuel type.')
plt.show()
```

### 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
df1 = df[['brand','vehicleType','price']]
```

```python
sns.set_theme(palette='pastel')
plt.figure(figsize=(30,20))
heat_table = df1.pivot_table(values='price',index='brand',columns='vehicleType',aggfunc=np.median)
sns.heatmap(heat_table,cmap='Greens',annot=True)
plt.title('Heatmap for Average price of a vehicle by brand as well as vehicle type')
plt.show()
```
