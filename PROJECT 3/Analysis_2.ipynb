### 2. What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with bothnumerical and visualization analysis

```python
average_prices = df.groupby(['vehicleType','gearbox'])['price'].mean()
```

```python
average_prices = pd.DataFrame(data = average_prices).reset_index()
```

```python
average_price = average_prices[average_prices['vehicleType'] != 'andere']
```

```python
sns.barplot(x=average_price['vehicleType'],y=average_price['price'],hue=average_price['gearbox'],ci=None)
plt.title('Average price for vehicles based on the type of vehicle as well as on the type of gearbox')
plt.xticks(rotation=45)
plt.show()
```

```python
average_price_1 = average_prices[average_prices['vehicleType'] == 'andere']
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x=average_price_1['vehicleType'],y=average_price_1['price'],hue=average_price_1['gearbox'],ci=None)
plt.title('Average price for vehicles based on the type of vehicle as well as on the type of gearbox for Andere')
plt.show()
```
